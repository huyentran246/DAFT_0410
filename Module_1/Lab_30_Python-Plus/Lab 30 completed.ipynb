```python
import numpy as np

def standardize(col):
    return (col - np.mean(col))/np.std(col)
df["cgpa_std"]= standardize(df["cgpa"])
df["gre_std"] = standardize(df["gre_score"])
df["lor_std"] = standardize(df["lor"])
print(df)
```

```python
score_1 = 0.4 * df["cgpa_std"] + 0.3 * df["gre_std"] + 0.3 * df["lor_std"]
df["score_1"] = score_1
df
```

```python
score_2 = df[["cgpa_std", "gre_std", "lor_std"]].max(axis=1)
df["score_2"] = score_2
df
```

```python
df.loc[(df["score_1"] > 0.8), "decision_1"] = "True"  
df.loc[(df["score_1"] < 0.8), "decision_1"] = "False"
df.loc[(df["score_2"] > 0.8), "decision_2"] = "True"
df.loc[(df["score_2"] < 0.8), "decision_2"] = "False"
df
```

```python
x = len(df[df['decision_1'] == 'True']) + len(df[df['decision_2'] == 'True'])
print("The number of applicants will be accepted to the program: ",x)
```
