```python
mean_year = int(data['BirthYear'].mean())
mean_year
```

```python
data["BirthYear"] = data["BirthYear"].fillna(mean_year)
data
```

```python
data_find = data.loc[(data["Risk"]=="High") | (data["Risk"]=="Medium")]
data_find
```

```python
data_find.groupby('Risk')["Profession"].value_counts()
```

```python
#data = data.apply(lambda x: x=="HR" if data["Profession"].isna() and data["Risk"]=="High" else x=="Manager")

data["Profession"] = data["Profession"].fillna(value=data["Risk"].apply(lambda x: x=="HR" if data["Risk"]=="High" else x=="Manager"))
data
```
