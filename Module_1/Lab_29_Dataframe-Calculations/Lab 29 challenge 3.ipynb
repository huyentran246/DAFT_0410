```python
import pandas as pd
import numpy as np

sales = pd.read_csv('/Users/valaba/Desktop/Orders.csv')
sales.head()
```

```python
# Q1: How to identify VIP & Preferred Customers?

orders = sales.groupby("CustomerID").sum("amount_spent")
orders
```

```python
vip_customers = orders[orders['amount_spent'] > orders['amount_spent'].quantile(0.95)]

vip_customers
```

```python
preferred_customers = orders[(orders['amount_spent'] > orders['amount_spent'].quantile(0.75)) & (orders['amount_spent'] <= orders['amount_spent'].quantile(0.95))]

preferred_customers
```

```python
# Q2: How to identify which country has the most VIP Customers?

vip_data = sales.merge(vip_customers, how='right', on="CustomerID")
vip_data
```

```python
vip_country = vip_data.groupby('Country').agg({'CustomerID':'count'})
vip_country.sort_values("CustomerID", ascending = False)

# UK has the highest number of VIP customers.
```

```python
# Q3: How to identify which country has the most VIP+Preferred Customers combined?

preferred_data = sales.merge(preferred_customers, how='right', on="CustomerID")

preferred_country = preferred_data.groupby('Country').agg({'Number_CustomerID':'count'})

preferred_country.sort_values("CustomerID", ascending = False)
```

```python
# Visually we can see the country who has the most VIP+Preferred Customers combined is UK.
# Let's try merge again to check:

vip_preferred = vip_country.merge(preferred_country, on="Country", how="inner", suffixes=("_vip","_preferred"))

vip_preferred["vip_preferred_combined"] = vip_preferred["CustomerID_vip"] + vip_preferred["CustomerID_preferred"]

vip_preferred.sort_values("vip_preferred_combined", ascending = False)

# True, the country who has the most VIP+Preferred Customers combined is UK.
```
