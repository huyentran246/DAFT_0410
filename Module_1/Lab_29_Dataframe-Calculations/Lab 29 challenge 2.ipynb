```python
import pandas as pd
import numpy as np

data = pd.read_csv('/Users/valaba/Desktop/Pokemon.csv')
data.head()
```

```python
# Problem Solving Iteration 1

sample = data.sample(10)

sample['computed_total'] = sample['HP'] + sample['Attack'] + sample['Defense'] + sample['Sp. Atk'] + sample['Sp. Def'] + sample['Speed']

sample['equal'] = sample['Total'] == sample['computed_total']

print(sample[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total', 'computed_total', 'equal']])
```

```python
# Problem Solving Iteration 2

data["Type 1"].unique()
```

```python
data["Type 2"].unique()
```

```python
data_types = pd.get_dummies(data['Type 1']) + pd.get_dummies(data['Type 2'])
data_types
```

```python
# Problem Solving Iteration 3

data_types = data_types.groupby(level=0, axis=1).sum()

data_encoded = pd.concat([data['Total'], data_types], axis=1)

correlation = data_encoded.corr()

correlation
```

```python
print("The no.1 pokemon type that is most likely to have the highest Total is: ",correlation.index[1])
```
