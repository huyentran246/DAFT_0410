```python
data.dropna(subset=['director_name','country','movie_title','genres', 'duration', 'color', 'director_facebook_likes', 'language','actor_3_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes'],inplace=True)
data.shape
```

```python
# 2. Use crosstabs to show the number of moovies per country

pd.crosstab([data['country']], data.index, values=data['movie_title'], aggfunc=len, margins=True, margins_name='Number of movies')[['Number of movies']][0:-1]
```

```python
# 3. Create a crosstab to show the number of films per director and genre splitted by country

x = pd.crosstab([data['country'], data['director_name'], data["genres"]], data.index, margins=True, margins_name='No of movies').replace(np.nan,0)
x["No of movies"].to_frame().sort_values(by="country", ascending=True)
```

```python
# 4. Average duration of filmes grouped by color

pd.crosstab(data["color"],data.index, values = data["duration"], aggfunc=np.mean, margins=True, margins_name='Average Duration')[['Average Duration']][0:-1]
```

```python
# 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

pd.crosstab([data["director_name"]], data.index, values=data["actor_1_facebook_likes"]+ data["actor_2_facebook_likes"] +data["actor_3_facebook_likes"], aggfunc=np.mean, margins=True, margins_name='Average likes')[['Average likes']]
```
