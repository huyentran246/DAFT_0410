```python
import numpy as np
import matplotlib.pyplot as plt

#Challenge 1

x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig, ax = plt.subplots()
ax.set(xlabel="X", ylabel="Y", title="Example Line Chart")
plt.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set(xlabel="X", ylabel="Z", title="Example Line Chart")
plt.plot(x,z)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set(title="Box Plot/Box and Whiskers", ylabel="Y")
plt.boxplot(y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set(title="Box Plot/Box and Whiskers", ylabel="Y")
plt.boxplot(z)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(20,5))

ax1.set_title("Beautiful day")
ax1.set_xlabel("Sunshine")
ax1.set_ylabel("Mood")
ax1.plot(x,y)

ax2.set(title="Rainy day", xlabel="Rain mesurement", ylabel="Romantic")
ax2.bar(x,y)

ax3.set(title="Busy day", xlabel="Mood", ylabel="Work")
ax3.scatter(x,y)

plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(30,10))

ax1.set_title("Example Line Chart")
ax1.set_xlabel("X")
ax1.set_ylabel("Z")
ax1.plot(x,z)

ax2.set(title="Example Bar Chart", xlabel="X", ylabel="Z")
ax2.bar(x,z)

ax3.set(title="Example Scatter Chart", xlabel="X", ylabel="Z")
ax3.scatter(x,z)

plt.show()
```

```python
# Now you will have to plot both 
# Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
# Hint: Use set_xscale and set_yscale. 

x = np.arange(0,100)    
y = x**2

fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, y, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, y, x, np.exp(x))
axes[1].set_yscale("log") #the log transformation
axes[1].set_title("Logarithmic scale (y)");
```
