# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('/Users/talia/Desktop/ironhack_labs/DAFT_0410/module_2/Lab_2_ Plotting-Multiple-Data-Series/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
data1 = data.groupby('ItemType').agg({"RetailSales":'sum', "RetailTransfers":"sum", "WarehouseSales":'sum'}).reset_index()

plt.bar(data1["ItemType"], data1["RetailSales"], label = "RetailSales")
plt.bar(data1["ItemType"], data1["RetailTransfers"], label = "RetailTransfers")
plt.bar(data1["ItemType"], data1["WarehouseSales"], label = "WarehouseSales")

plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.title('Total Sales by Item Type')

plt.xticks(rotation="vertical") # make x axis with words vertical.

plt.legend()
plt.show()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data["Total Sales"] = data["RetailSales"] + data["RetailTransfers"] + data["WarehouseSales"]
top = data.groupby("Supplier")["Total Sales"].sum().to_frame()
top_desc = top.sort_values(by="Total Sales", ascending=False)
top_desc
```

```python
top10 = top_desc.iloc[0:11]
top10
```

```python
top10.plot(kind="barh", color="pink", title='Sales of Top 10 Suppliers')
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
sales_by_month = data.groupby("Month")[["RetailSales","RetailTransfers","WarehouseSales"]].mean()
#retailtransfers_month = data.groupby("Month")["RetailTransfers"].mean()
#warehousesales_month = data.groupby("Month")["WarehouseSales"].mean()
sales_by_month.plot(kind="line");

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Average Sales per Month')
```

## 4. Plot the same information as above but as a bar chart.

```python
sales_by_month.plot(kind="bar");

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Average Sales per Month')
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
data["date"] = pd.to_datetime(data[["Year","Month"]].assign(Day=1))
data.set_index("date", inplace=True)

retailsales_by_item = data.groupby('ItemType')["RetailSales"].resample('M').sum().unstack('ItemType')

retailsales_by_item.plot(kind="line", title="Total Retail Sales by Item Type over time",xlabel="Date",ylabel='Sales');
```

## 6. Plot the same information as above but as a bar chart.

```python
retailsales_by_item.plot(kind="bar", title="Total Retail Sales by Item Type over time",xlabel="Date",ylabel='Sales');
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.scatterplot(x="RetailSales", y="RetailTransfers", data=data, hue="ItemType");
```
